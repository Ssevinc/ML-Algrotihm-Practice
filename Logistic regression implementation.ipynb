**This notebook contains a very simple implementation of logistic regression for breast cancer prediction with a model accuracy of 97%.**

**Here is some of the key steps in the code:**

    1. Import necessary libraries.
    2. Load and preprocess the dataset.
    3. Split data into train/test sets.
    4. Standardize features.
    5. Train logistic regression.
    6. Evaluate performance (accuracy, confusion matrix).
    7. Apply a simple test from the test dataset.


    

```python

import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
breast_cancer = load_breast_cancer()
x,y =breast_cancer.data,breast_cancer.target
#y_binary = (y > np.median(y)).astype(int) this is required if the output needs to be transformed into two classes rather than a quantitative value but no need for it in this dataset

print("target names:",breast_cancer.target_names)
print("feature names:",breast_cancer.feature_names)
```

```python
df = pd.DataFrame(data=breast_cancer.data,columns=breast_cancer.feature_names)
df.head()
```

```python
x_train, x_test, y_train,y_test = train_test_split(
    x,y, test_size= 0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
patient_data = x_test[0].reshape(1, -1)  # Reshape it to 2D for the model

# Make the prediction
prediction = model.predict(patient_data)


if prediction == 0:
    print("The model predicts the tumor is: Malignant")
else:
    print("The model predicts the tumor is: Benign")
```
