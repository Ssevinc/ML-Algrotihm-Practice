A naive bayes implementation using scikit learn with my small 50 datapoints dataset

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/kaggle/input/whether-a-message-is-a-scam-or-not/dataset.csv')
x = df[['free', 'gift', 'investment', 'bonus']]
y = df['label']

X_train,X_test,y_train,y_test = train_test_split(
    x,y, test_size=0.25,random_state=2
)
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
print(y_pred)
print("Number of mislabeled points out of a total %d points : %d"
      % (X_test.shape[0], (y_test != y_pred).sum()))

```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:", accuracy)
```
